# Análise estatística utilizando o Jupyter

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv('arq_infracoes_bsb21.csv')
```

```python
dados.cometimento = pd.to_datetime(dados.cometimento, format='%d/%m/%Y')
```

```python
dias_pt = {0:'Segunda',
           1:'Terça',
           2:'Quarta',
           3:'Quinta',
           4:'Sexta',
           5:'Sábado',
           6:'Domingo'}
dados['dias_da_semana'] = dados.cometimento.dt.weekday.map(dias_pt)
```

```python
horas_fim_de_semana = dados[(dados.dias_da_semana == 'Sábado') | 
                            (dados.dias_da_semana == 'Domingo')][
                                ['hora_cometimento','dias_da_semana','tipo_infracao', 'descricao']]
```

```python
horas_fim_de_semana[['Hora', 'Minutos']] = horas_fim_de_semana.hora_cometimento.str.split(':', expand=True)
```

```python
graf1 = horas_fim_de_semana.Hora.value_counts().sort_index()
graf2 = horas_fim_de_semana.query("tipo_infracao == '7579-0'").Hora.value_counts().sort_index()
```

```python
fig, axs = plt.subplots(2,sharex=True, figsize=(14,8))
plt.subplots_adjust(wspace=0.25, hspace=0.25)

axs[0].plot(graf1.index, graf1.values,c='navy', )
axs[0].set_title('Autos registrados nos finais de semana')
axs[0].set_ylabel('Autos registrados') 

axs[1].plot(graf2.index, graf2.values,c='tomato')
axs[1].set_title('Registros de recusa da realização do exame nos finais de semana')
axs[1].set_ylabel('Autos registrados')
axs[1].set_xlabel('Horário')

plt.savefig('Gráficos Autos finais de semana e negar-se.jpeg')
plt.show()
```

```python
graf3 = graf1[0:7]
graf4 = graf2[0:7]
```

```python
fig, axs = plt.subplots(2,sharex=True, figsize=(14,8))
plt.subplots_adjust(wspace=0.25, hspace=0.25)

axs[0].plot(graf3.index, graf3.values,c='navy' )
axs[0].set_title('Autos registrados nas madrugadas dos finais de semana')
axs[0].set_ylabel('Autos registrados')

axs[1].plot(graf4.index, graf4.values,c='tomato')
axs[1].set_title('Registros de recusa da realização do exame nas madrugadas dos finais de semana')
axs[1].set_ylabel('Autos registrados')
axs[1].set_xlabel('Horário')

plt.savefig('Gráficos Autos finais de semana e negar-se durante a madrugada.jpeg')
plt.show()
```

```python
autos_fds = graf3.to_frame('total')
autos_fds['negar_exame'] = graf4
correlacao = autos_fds.corr('pearson')
correlacao
```

```python
plt.subplots(figsize=(14,8))
ax = sns.heatmap(correlacao, annot = True, fmt=".1f", annot_kws={'size':30}, cmap='OrRd')
ax.set_title('Correlação entre negar exame e total final de semana durante a madrugada')
plt.savefig('Correlação negar exame e total final de semana durante a madrugada.jpeg')
plt.show()
```

```python
autos_fds = graf1.to_frame('total')
autos_fds['negar_exame'] = graf2
correlacao = autos_fds.corr('pearson')
correlacao
```

```python
plt.subplots(figsize=(14,8))
ax = sns.heatmap(correlacao, annot = True, fmt=".1f", annot_kws={'size':30}, cmap='OrRd')
ax.set_title('Correlação entre negar exame e total final de semana')
plt.savefig('Correlação negar exame e total final de semana.jpeg')
plt.show()
```

```python
top10 = list(horas_fim_de_semana.tipo_infracao.value_counts().head(10).index)
infracoes = pd.DataFrame()
for i in top10:
    total = horas_fim_de_semana[horas_fim_de_semana.tipo_infracao == i].Hora.value_counts().sort_index()
    infracoes[i] = total
```

```python
infracoes['total'] = graf1.values
```

```python
correlacao = infracoes.corr()
```

```python
plt.subplots(figsize=(14,8))
ax = sns.heatmap(correlacao, annot = True, fmt=".1f", annot_kws={'size':18}, cmap='OrRd')
ax.set_title('Correlação entre todas infrações e total final de semana')
plt.savefig('Correlação entre todas infrações e total final de semana.jpeg')
plt.show()
```
